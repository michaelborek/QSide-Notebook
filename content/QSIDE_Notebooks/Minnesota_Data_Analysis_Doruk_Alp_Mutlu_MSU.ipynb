```python
fp = open("allmn.csv", "r")
fp.readline()
sdict = {}
for line in fp:
  line = line.split(",")
  row1 = [str(line[0]) , str(line[1]), str(line[2]), str(line[11]) , str(line[12]), str(line[14]), str(line[30]), str(line[31]), str(line[38]), str(line[39]), str(line[40]), str(line[49])]
  line[7] = int(line[7])
  line[8] = int(line[8])
  if int(row1[5]) not in sdict:
    sdict[int(row1[5])] = 1
  if line[7] == 1 and line[8] == 1:
    writer_csv_male_white.writerow(row1)
  if line[7] == 1 and line[8] == 2:
    writer_csv_male_black.writerow(row1)
  if line[7] == 1 and line[8] == 3:
    writer_csv_male_am_ind.writerow(row1)
  if line[7] == 1 and line[8] == 4:
    writer_csv_male_hispanic.writerow(row1)
  if line[7] == 1 and line[8] == 5:
    writer_csv_male_asian.writerow(row1)
  if line[7] == 2 and line[8] == 1:
    writer_csv_female_white.writerow(row1)
  if line[7] == 2 and line[8] == 2:
    writer_csv_female_black.writerow(row1)
  if line[7] == 2 and line[8] == 3:
    writer_csv_female_am_ind.writerow(row1)
  if line[7] == 2 and line[8] == 4:
    writer_csv_female_hispanic.writerow(row1)
  if line[7] == 2 and line[8] == 5:
    writer_csv_female_asian.writerow(row1)
print(sorted(sdict))
#sex is either 1(male) or 2(female)
#race 1White, 2 Black, 3 Am Ind, 4 Hisp, 5Asian, 6 Other, 7 Unknown
#ethnic 0 none, 1 hispanic, 2 unknown
#Ideas, checking how different races might have got different sentences with similar background and crime' severity levels

```
