```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
minn_data = pd.read_csv('/content/drive/MyDrive/allmn.csv')
```

```python
minn_data
```

```python
sev_crime_data = minn_data[["severity", "subwpn", "subsex", "subdrug"]]
sev_crime_data
```

```python
fig, axes = plt.subplots(1, 3)
sns.scatterplot(sev_crime_data["subwpn"], sev_crime_data["severity"], ax=axes[0])
sns.scatterplot(sev_crime_data["subsex"], sev_crime_data["severity"], ax=axes[1])
sns.scatterplot(sev_crime_data["subdrug"], sev_crime_data["severity"], ax=axes[2])
plt.tight_layout()
```

```python
jail_sentence_data = minn_data[["race", "lifeminmonths", "condconf"]]
jail_sentence_data
```

```python
fig, axes = plt.subplots(1, 3)
sns.scatterplot(jail_sentence_data["race"], jail_sentence_data["lifeminmonths"], hue=sev_crime_data["subwpn"], ax=axes[0])
sns.scatterplot(jail_sentence_data["race"], jail_sentence_data["lifeminmonths"], hue=sev_crime_data["subsex"], ax=axes[1])
sns.scatterplot(jail_sentence_data["race"], jail_sentence_data["lifeminmonths"], hue=sev_crime_data["subdrug"], ax=axes[2])
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 3)
sns.scatterplot(jail_sentence_data["lifeminmonths"], jail_sentence_data["condconf"], hue=sev_crime_data["subwpn"], ax=axes[0])
sns.scatterplot(jail_sentence_data["lifeminmonths"], jail_sentence_data["condconf"], hue=sev_crime_data["subsex"], ax=axes[1])
sns.scatterplot(jail_sentence_data["lifeminmonths"], jail_sentence_data["condconf"], hue=sev_crime_data["subdrug"], ax=axes[2])
plt.tight_layout()
```
